```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import re

```

```python
# Baca CSV
df = pd.read_csv('summary.csv')

# Tampilkan dataframe awal
print("Dataframe Awal:")
df
```

```python
# Case folding
df['Summary'] = df['Summary'].str.lower()

# Menampilkan dataframe setelah case folding
print("\nDataframe Setelah Case Folding:")
df
```

```python
# Fungsi untuk menghapus tanda baca
def remove_punctuation(text):
    return ''.join([char for char in text if char not in string.punctuation])

# Hapus tanda baca dari kolom 'ringkasan'
df['Summary'] = df['Summary'].apply(remove_punctuation)

# Tampilkan dataframe setelah menghapus tanda baca
print("\nDataframe Setelah Menghapus Tanda Baca:")
df
```

```python
def remove_numbers(text):
    return ''.join([char for char in text if not char.isdigit()])
# Hapus angka dari kolom 'Summary' yang sudah tidak memiliki tanda baca
df['Summary'] = df['Summary'].apply(remove_numbers)
print("\nDataframe Setelah Menghapus Angka:")
df
```

```python
# Daftar stop words dalam bahasa Indonesia
stop_words = set(stopwords.words('indonesian'))
stop_words.discard('bapak')  # Hapus 'bapak' dari daftar stop words
stop_words.discard('ibu')    # Hapus 'ibu' dari daftar stop words

def remove_stopwords(text):
    words = word_tokenize(text)
    words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(words)

# Terapkan fungsi remove_stopwords pada kolom 'Summary' dari DataFrame 'df'
df['Summary'] = df['Summary'].apply(remove_stopwords)
df
#stop_words
```

```python
# Fungsi untuk menghapus spasi dari kata-kata yang diberikan
def remove_space(text, words_to_remove):
    for original, replacement in words_to_remove.items():
        text = text.replace(original, replacement)
    return text

# Membaca file CSV yang berisi daftar kata-kata yang akan diproses
words_df = pd.read_csv('remove_space.csv')
words_to_remove = dict(zip(words_df['original'], words_df['replacement']))

# Hapus spasi dari kata-kata terkait dari kolom 'Summary'
df['Summary'] = df['Summary'].apply(lambda text: remove_space(text, words_to_remove))
df
```

```python
# Fungsi untuk menghapus kata-kata yang tidak relevan
def remove_word(text, words_to_remove):
    for word in words_to_remove:
        pattern = r'\b' + re.escape(word) + r'\b'
        text = re.sub(pattern, '', text)
    return text

# Membaca file CSV yang berisi daftar kata yang akan dihapus
words_df = pd.read_csv('list_hapus_kata.csv')
detikcom_related_words = words_df['hapus_kata'].tolist()

# Hapus kata-kata terkait detikcom dari kolom 'Summary'
df['Summary'] = df['Summary'].apply(lambda text: remove_word(text, detikcom_related_words))
df
```

```python
# Fungsi untuk menghapus nama orang
def remove_word(text, words_to_remove):
    for word in words_to_remove:
        pattern = r'\b' + re.escape(word) + r'\b'
        text = re.sub(pattern, '', text)
    return text

# Membaca file CSV yang berisi daftar kata yang akan dihapus
words_df = pd.read_csv('list_hapus_nama.csv')
detikcom_related_words = words_df['hapus_nama'].tolist()

# Hapus kata-kata terkait detikcom dari kolom 'Summary'
df['Summary'] = df['Summary'].apply(lambda text: remove_word(text, detikcom_related_words))

df
```

```python
# Fungsi untuk membaca kata-kata dari file CSV
def read_words_from_csv(file_path):
    df = pd.read_csv(file_path)
    words_set = set(df['word'])
    return words_set

# Baca kata-kata dari file CSV
words_set = read_words_from_csv('not_remove.csv')

# Fungsi untuk menghilangkan sufiks 'nya' di akhir kata
def remove_suffix_nya(words):
    return [word[:-3] if word.endswith('nya') else word for word in words]

# Fungsi untuk preprocessing teks
def preprocess_text(text, words_set):
    # Tokenisasi teks menjadi kata-kata
    words = nltk.word_tokenize(text)
    # Hapus kata-kata yang hanya terdiri dari satu huruf
    filtered_words = [word for word in words if len(word) > 6 or word in words_set]
    # Hapus sufiks 'nya' dari setiap kata
    filtered_words = remove_suffix_nya(filtered_words)
    # Gabungkan kata-kata yang tersaring kembali menjadi teks
    filtered_text = ' '.join(filtered_words)
    return filtered_text

# Pra-pemrosesan teks pada kolom 'Summary'
df['Summary'] = df['Summary'].apply(lambda x: preprocess_text(x, words_set))
df
```

```python
# Fungsi untuk menghapus kata yang diulang dalam satu kalimat
def remove_repeated_words(text):
    words = text.split()
    seen = set()
    result = []
    for word in words:
        if word not in seen:
            result.append(word)
            seen.add(word)
    return ' '.join(result)

# Hapus kata yang diulang pada kolom 'Summary'
df['Summary'] = df['Summary'].apply(remove_repeated_words)
df
```

```python
# Simpan DataFrame ke dalam file CSV
df.to_csv('pra-pemrosesan.csv', index=False)

print("Hasil akhir disimpan dalam 'hasil_proses.csv'")
```
